```
import numpy as np
import matplotlib.pyplot as plt

def smooth(img):
  top = img[:-2,1:-1]
  left = img[1:-1, :-2]
  right = img[1:-1,2:]
  bottom = img[2:,1:-1]
  center = img[1:-1,1:-1]
  blurred = (top+left+right+bottom+center)/5.0
  return blurred

avg_img = img

for i in range(50):
  avg_img = smooth(avg_img)

plt.figure()
plt.imshow(avg_img, cmap=plt.cm.hot)
plt.show()
```

```
from numpy import loadtxt

data = loadtxt("Numpy-Tutorial-SciPyConf-2019/exercises/wind_statistics/wind.data")

print(data.shape)
```

```
#Calculate the windspeed min, max, mean, and standard deviation for each location in a single array

dates = data[:, :3 ]
winds = data[:,  3:]  #Slice does not make new object in memory, convenient and efficient!

stats = ([winds.min(axis=0)],
         [winds.max(axis=0)],
         [winds.mean(axis=0)],
         [winds.std(axis=0)])

# 2D array of days (rows) and stats (columns)
dailystats = ([winds.min(axis=1)],
              [winds.max(axis=1)],
              [winds.mean(axis=1)],
              [winds.std(axis=1)])

whereMax = winds.argmax(axis=1)
#print(whereMax)
#whenMax
#np.unravel_index(winds.argmax().winds.shape)
#print(np.where(winds == winds.max()))
maxSpeedIndex = winds.max(axis=1).argmax()
maxSpeedDate = dates[maxSpeedIndex]
print("The greatest windspeed was recorded on: ")
print(maxSpeedDate[1], ", ", maxSpeedDate[2], ", 19", maxSpeedDate[0])

#Only January Speeds
#boolean array of januaries
jan = dates[:,1] == 1
janAVG = winds[jan].mean(axis=0)
```

```
import matplotlib.pyplot as plt

plt.figure()
plt.plot(janAVG, 'r')
plt.show()
```
